### Exercise 2

```python
def selection_sort(lst):
    for i in range(len(lst)):
        min_index = i
        for j in range(i+1, len(lst)):
            if lst[min_index] > lst[j]:
                min_index = j
        lst[i], lst[min_index] = lst[min_index], lst[i]
```

```python
selection_sort(police_camera)
police_camera
```

### Exercise 3

```python
def binary_search(lst, left, right, x):
    if right >= left:
        middle = (left + right) // 2
        if lst[middle] == x:
            return middle
        elif lst[middle] < x:
            return binary_search(lst, middle+1, right, x)
        else:
            return binary_search(lst, left, middle-1, x)
    else:
        return -1
```

```python
binary_search(police_camera, 0 , len(police_camera)-1 , 1)
```

```python
def find_middle(lst):
    arr = np.sort(np.array(lst))
    mid =len(arr)//2
    if len(arr) % 2 == 0:
        return arr[mid-1]
    return arr[mid]
```
